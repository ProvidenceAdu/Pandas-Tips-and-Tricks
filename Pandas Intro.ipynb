#### Merging Columns

- Like join in ArcGIS, pandas allows you to join two datasets

- There is a key difference here, in pandas, Joins are done with the Merge function

- Although pandas has a Join function, it only joins data using the index instead of a common ID field


-  pd.merge(left, right, how="left", on=["ID"]), If you have one common ID

-  pd.merge(left, right, how="left", on=["ID1", "ID2"]), If you two common different IDs


```python
NCBelowPoverty =  pd.read_csv("NCBelowPovertyLine.csv")
NCEducation = pd.read_csv("NCEducation.csv")
```

```python
NCBelowPoverty
```

```python
NCEducation
```

```python
NCcombined = pd.merge(NCBelowPoverty, NCEducation, how = 'left', on = ["GEOID"])
NCcombined
```

```python
NCcombined.columns
```

#### Grouping and Sorting Data 

- Pandas allows you to group data categories and undertake computations such mean, or count 

- Pandas also allows you to sort column either ascending or descending 

```python
webcsv = pd.read_csv("https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv")
webcsv
```

```python
## Group data by contient and check their mean 

continentmean = webcsv.groupby(['continent']).mean()
continentmean 
```

```python
#Group data by content and check their count 

webcsv.groupby(['continent']).size()
```

```python
### Sort the gdp by continent 

continentmean.sort_values(by =['gdpPercap'], ascending= False) 
```

#### Handling null values

- You can drop null values using the inbuilt function dropna()


- You can also replace null values with specific numbers using the inbuilt function fillna()


- In order for fillna() or dropna() function to stay in memory, you have to map it to a new variable



- Caution 
    -  Dropping null values deletes all rows which have null values
    -  Filling null values replaces all null values with the specified value


```python
Davidson = pd.read_csv("Davidson.txt")
Davidson
```

```python
Davidson.dropna(inplace= True)
```

```python
Davidson
```

```python
Davidson.fillna(20)
```

```python
#Not inplace, if you want the fill to overwrite the existing dataframe, you have to set inplace = True
Davidson
```
